###Task 2 — Clean the Data

```python
#drop rows where target is null
df = df.dropna(subset=['selling_price'])
df
```

```python
#fill missing values
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
df = df[numeric_cols].fillna(df[numeric_cols].mean())
df
```

```python
#clean brand column
df['brand'] = df['brand'].str.strip().str.lower()

#extract numeric from mileage
df['mileage'] = df['mileage'].astype(str)

df['mileage'] = df['mileage'].str.extract('([0-9.]+)', expand=False)

df['mileage'] = pd.to_numeric(df['mileage'], errors='coerce')

#remove duplicates
df = df.drop_duplicates()
df
```

###Task 3 — Compute Baseline MAE



```python
from sklearn.metrics import mean_absolute_error

#mean value for selling price
mean_price = df['selling_price'].mean()

#predictions
pred = [mean_price] * len(df)

#actual values
actul = df['selling_price']

#mean absolute error
mean_absolut = mean_absolute_error(actul, pred)

print("Baseline MAE:", mean_absolut)
```
